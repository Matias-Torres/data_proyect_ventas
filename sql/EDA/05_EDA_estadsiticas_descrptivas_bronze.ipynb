## **EDA - Estadísticas Descriptivas**


```python
%sql
-- ESTADISTICA EN GENERAL
SELECT
  COUNT(DISTINCT venta) AS total_tickets,
  SUM(cant) AS total_unidades,
  ROUND(SUM(total),2) AS facturacion_total,
  ROUND(AVG(total),2) AS ticket_promedio
FROM ventas_EDA
WHERE vtaestado = 'normal'
```

```python
%sql
-- ESTADISTICAS DE PRECIO SEGUN PRODUCTOS
SELECT
  descrip,
  COUNT(*) AS total_ventas,
  ROUND(MIN(precio),2) AS precio_minimo,
  ROUND(MAX(precio),2) AS precio_maximo,
  ROUND(AVG(precio),2) AS precio_promedio,
  ROUND(PERCENTILE(precio,0.5),2)  AS precio_mediana,
  ROUND(PERCENTILE(precio,0.1),2) AS precio_p1,
  ROUND(PERCENTILE(precio,0.9),2) AS precio_p90
FROM ventas_EDA
WHERE vtaestado='normal'
 AND precio IS NOT NULL 
 AND precio > 0
 AND precio != 0.01
 AND descrip NOT LIKE 'envio%'
GROUP BY descrip
ORDER BY total_ventas DESC
```

```python
%sql
-- ESTADISTCA DE CANTIDADES
SELECT
  COUNT(*) AS total_ventas,
  ROUND(MIN(cant),2) AS cant_minimo,
  ROUND(MAX(cant),2) AS cant_maximo,
  ROUND(AVG(cant),2) AS cant_promedio,
  ROUND(PERCENTILE(cant,0.5),2)  AS cant_mediana
FROM ventas_EDA
WHERE vtaestado='normal'
```

```python
%sql
-- ESTADISTICAS DE INGRESOS 
SELECT
  COUNT(*) AS total_ventas,
  ROUND(SUM(total),2) AS ingreso_total,
  ROUND(MIN(total),2) AS ingreso_minimo,
  ROUND(MAX(total),2) AS ingreso_maximo,
  ROUND(AVG(total),2) AS ingreso_promedio,
  ROUND(PERCENTILE(total,0.5),2)  AS ingreso_mediana,
  ROUND(PERCENTILE(total,0.1),2) AS ingreso_p1,
  ROUND(PERCENTILE(total,0.9),2) AS ingreso_p90
FROM ventas_EDA
WHERE vtaestado='normal'
 AND total IS NOT NULL 
 AND total > 0
 AND total != 0.01
ORDER BY ingreso_total DESC
```
