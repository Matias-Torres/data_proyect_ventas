## **EDA - Cardinalidad y Distribución de Categóricos**


```python
%sql
-- DISTRIBUCION DE VTAOPERACION
SELECT
    vtaoperacion,
    COUNT(*) AS total,
    ROUND(COUNT(*) * 100.0 / SUM(COUNT(*)) OVER(),2) AS pct_vtaoperacion
FROM catalog_ventas.raw.ventas_heladeria_bronze
    GROUP BY vtaoperacion
    ORDER BY total DESC

/*
VL: venta a socios 
VF: venta a no socios
VG: venta gastronomica
CL: otros
*/

/*
La columna presenta baja cardinalidad, por lo que es adecuada para modelarse como dimensión categórica en el modelo analítico.

No se detectan valores anómalos ni categorías inesperadas, lo que indica buena consistencia del dato.

Las categorías minoritarias deben preservarse, ya que pueden representar tipos de operación relevantes para análisis específicos del negocio.

*/
```

```python
%sql

-- DISTRIBUCION DE VTAESTADO
SELECT
    vtaestado,
    COUNT(*) AS total,
    ROUND(COUNT(*) * 100.0 / SUM(COUNT(*)) OVER(),2) AS pct_vtaestado
FROM catalog_ventas.raw.ventas_heladeria_bronze
    GROUP BY vtaestado
    ORDER BY total DESC


/*
La gran mayoría de las transacciones corresponden a ventas Normales, representando 97.16% del total, mientras que las ventas Anuladas representan solo 2.84%.

Esto indica que:

El dataset presenta alta consistencia operativa, con una baja proporción de cancelaciones.

La tasa de anulaciones se mantiene en un rango esperable para sistemas de punto de venta (POS)

Las ventas anuladas deben excluirse de los cálculos de KPIs, ya que no representan ingresos reales.
*/
```

```python
%sql

-- DISTRIBUCION DE CLINOMBRE
SELECT
    clinombre,
    COUNT(*) AS total,
    ROUND(COUNT(*) * 100.0 / SUM(COUNT(*)) OVER(),2) AS pct_clinombre
FROM catalog_ventas.raw.ventas_heladeria_bronze
    GROUP BY clinombre
    ORDER BY total DESC

/*
El 95% corresponde a la venta minorista en sucursales.
El 5% restante corresponde a ventas a comerciantes o otros clientes.
*/
```

```python
%sql
-- DISTRIBUCION DE CONDVTAPOS
SELECT
    condvtapos,
    COUNT(*) AS total,
    ROUND(COUNT(*) * 100.0 / SUM(COUNT(*)) OVER(),2) AS pct_condvtapos
FROM catalog_ventas.raw.ventas_heladeria_bronze
    GROUP BY condvtapos
    ORDER BY total DESC

/*
TD: tarjeta debito
EF: efectivo
TC: tarjeta credito
MU: multiples opciones de pagos
XX: cancelados
*/

/*
La distribución de condiciones de venta muestra que los pagos electrónicos predominan en las transacciones, representando el 65,49% del total, con la tarjeta de débito como medio de pago principal (36,88%). El efectivo mantiene una participación significativa (31,06%), lo que indica que continúa siendo un método relevante en operaciones minoristas
*/
```

```python
%sql
-- DISTRIBUCION DE DELIVERY
SELECT
    delivery,
    COUNT(*) AS total,
    ROUND(COUNT(*) * 100.0 / SUM(COUNT(*)) OVER(),2) AS pct_delivery
FROM catalog_ventas.raw.ventas_heladeria_bronze
    GROUP BY delivery
    ORDER BY total DESC

/*
El análisis de la variable delivery muestra que la mayoría de las ventas se realizan sin servicio de entrega a domicilio (82,69%), mientras que el delivery representa el 17,31% del total de operaciones. Esto indica que el negocio se apoya principalmente en el consumo presencial, aunque también cuenta con un canal de venta online a través del servicio de delivery
*/
```

```python
%sql
-- DISTRIBUCION DE ARTICULO
SELECT
    articulo,
    descrip,
    COUNT(*) AS total,
    ROUND(COUNT(*) * 100.0 / SUM(COUNT(*)) OVER(),2) AS pct_articulo
FROM catalog_ventas.raw.ventas_heladeria_bronze
    GROUP BY articulo, descrip
    ORDER BY total DESC

/*
Distribución de ventas altamente concentrada en los formatos de helado por peso (1/4 kg, 1/2 kg y 1 kg), los cuales representan una proporción significativa del total de transacciones. Asimismo, se observa una participación relevante de productos individuales como palitos y bombones helados, asociados a consumo inmediato.
*/
```

```python
%sql
-- DISTRIBUCION DE CLIENTE 
SELECT
    cliente,
    clinombre,
    COUNT(*) AS total,
    ROUND(COUNT(*) * 100.0 / SUM(COUNT(*)) OVER(),2) AS pct_descrip
FROM catalog_ventas.raw.ventas_heladeria_bronze
    GROUP BY cliente,clinombre
    ORDER BY total DESC

/*
La distribución de clientes muestra una alta concentración en ventas a consumidor final (94,63%), mientras que el resto de los clientes presentan participaciones individuales muy bajas,
*/
```
