## **EDA - Analisis mas avanzado**


```python
%sql
--TOP DE 10 ARTICULOS POR PRECIO PROMEDIO
WITH precio_promedio_articulo AS (
  SELECT
    articulo,
    descrip,
    COUNT(*) AS total_ventas,
    ROUND(AVG(precio),2) AS precio_promedio,
    ROW_NUMBER() OVER (ORDER BY AVG(precio) DESC) AS rank
  FROM ventas_EDA
  WHERE vtaestado = 'normal'
    AND precio IS NOT NULL 
    AND precio > 0
    AND precio <> 0.01
  GROUP BY articulo, descrip
)

SELECT
  rank,
  articulo,
  descrip,
  precio_promedio,
  total_ventas
FROM precio_promedio_articulo
WHERE rank <= 10
ORDER BY rank;
```

```python
%sql
--FACTURACION TOTAL POR ARTICULO
SELECT
  articulo,
  descrip,
  SUM(total) AS facturacion_total
FROM ventas_EDA
WHERE vtaestado='normal'
 AND total IS NOT NULL 
 AND total != 0.01
 AND total > 0
GROUP BY articulo, descrip
ORDER BY facturacion_total DESC
LIMIT 10;
```
