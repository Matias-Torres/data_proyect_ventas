## **Dashboard con múltiples CTEs**

```python
%sql
--RESUMEN COMPLETO DEL DATASET
WITH metricas_generales AS (
  SELECT
    COUNT(*) AS total_registros,
    COUNT(DISTINCT venta) AS total_ventas,
    COUNT(DISTINCT articulo) AS catalogo_productos,
    SUM(cant) AS total_unidades,
    ROUND(AVG(total),2) AS ticket_promedio
  FROM ventas_EDA
),
problemas_calidad AS (
SELECT
  ROUND(SUM(CASE WHEN precio IS NULL OR precio < 0  AND precio !=0.1 THEN 1 ELSE 0 END) * 100.0 / COUNT(*), 2) AS pct_precio_invalido,
  ROUND(SUM(CASE WHEN total IS NULL OR total < 0  AND total !=0.1 THEN 1 ELSE 0 END) * 100.0 / COUNT(*), 2) as pct_total_invalido
FROM ventas_EDA
)
SELECT
  '📊 RESUMEN DEL DATASET' AS seccion ,
  m.total_registros,
  m.total_ventas,
  m.catalogo_productos,
  m.total_unidades,
  m.ticket_promedio,
  c.pct_precio_invalido,
  c.pct_total_invalido
FROM metricas_generales m, problemas_calidad c
```

## 📋 Hallazgos del EDA

### Problemas de calidad identificados:

1. **Datos invalidos:** La columna (precio,total) aparecen valores como 0.1
1. **Campos nulos:** Columna como comprobante tiene alto % de valores nulos
3. **Datos mixtos:** Nombre de articulos y tipos de envios mezclados (requiere normalización)
4. **Posibles duplicados:** Mismo articulo con diferentes descripciones
5. **Outliers:** Precios extremadamente altos o bajos que podrían ser errores
6. **Texto sucio:** La columna `descrip` tiene caracteres mal codificados
