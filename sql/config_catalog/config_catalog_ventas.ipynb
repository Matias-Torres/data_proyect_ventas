```python
%sql 
-- PASO 1 : CREAR CATALOGO
CREATE CATALOG IF NOT EXISTS catalog_ventas;

USE CATALOG catalog_ventas;
     
SHOW CATALOGS
```

```python
%sql
-- PASO 2: CREAR ESQUEMAS RAW-SILVER-GOLD
CREATE SCHEMA IF NOT EXISTS catalog_ventas.raw
COMMENT 'Esquema para datos crudos sin procesar(Bronze Layer)';

CREATE SCHEMA IF NOT EXISTS catalog_ventas.silver
COMMENT 'Esquema para datos limpios y de calidad (Silver Layer)';

CREATE SCHEMA IF NOT EXISTS catalog_ventas.gold
COMMENT 'Esquema para datos agregados y analisis de negocio (Gold Layer)';

SHOW SCHEMAS
```

```python
%sql
-- PASO 2: CREAR EL VOLUME PARA ARCHIVOS (ventas_helados_bronze.csv)
CREATE VOLUME IF NOT EXISTS catalog_ventas.raw.archivos
COMMENT 'Volume para almacenar archivos csv crudos';

SHOW VOLUMES IN catalog_ventas.raw
```
