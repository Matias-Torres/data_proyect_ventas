```python
%sql
INSERT INTO catalog_ventas.raw.ventas_heladeria_bronze
SELECT
  *,
  CURRENT_TIMESTAMP(),
  input_file_name() AS _source_file
FROM read_files(
  '/Volumes/catalog_ventas/raw/archivos/*.csv',
  format => 'csv',
  header => true,
  sep => ';'
)
WHERE input_file_name() NOT IN (
  SELECT _source_file FROM catalog_ventas.raw.ventas_heladeria_bronze
)
```
