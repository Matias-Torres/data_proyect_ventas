```python
%sql
MERGE INTO catalog_ventas.gold.dim_clientes AS tgt
USING (
  SELECT 
  DISTINCT 
    cliente
  FROM catalog_ventas.silver.ventas_clean
  WHERE cliente IS NOT NULL
) AS src
ON tgt.cliente = src.cliente

WHEN NOT MATCHED THEN
  INSERT (cliente, _created_at)
  VALUES (src.cliente,  CURRENT_TIMESTAMP())
```
