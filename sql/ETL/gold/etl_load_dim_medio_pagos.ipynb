```python
%sql
INSERT INTO catalog_ventas.gold.dim_medio_pago (medio_pago, _created_at)
SELECT 
DISTINCT 
    src.medio_pago, 
    CURRENT_TIMESTAMP()
FROM catalog_ventas.silver.ventas_clean src
WHERE src.medio_pago IS NOT NULL
AND NOT EXISTS (
    SELECT 1
    FROM catalog_ventas.gold.dim_medio_pago tgt
    WHERE tgt.medio_pago = src.medio_pago
);
```
