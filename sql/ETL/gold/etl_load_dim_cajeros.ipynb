```python
%sql
MERGE INTO catalog_ventas.gold.dim_cajero AS tgt
USING (
  SELECT 
  DISTINCT 
    usulogin,
    turno
  FROM catalog_ventas.silver.ventas_clean
  WHERE usulogin IS NOT NULL
) AS src
ON tgt.usulogin = src.usulogin

WHEN MATCHED AND COALESCE(tgt.turno,'') <> COALESCE(src.turno,'')
THEN UPDATE
  SET
      tgt.turno = src.turno,
      tgt._updated_at=CURRENT_TIMESTAMP()

WHEN NOT MATCHED THEN
  INSERT (usulogin, turno, _created_at)
  VALUES (src.usulogin, src.turno, CURRENT_TIMESTAMP());
```
