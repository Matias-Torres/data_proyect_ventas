```python
%sql
MERGE INTO catalog_ventas.gold.dim_cajero AS tgt
USING (
  SELECT DISTINCT usulogin
  FROM catalog_ventas.silver.ventas_clean
  WHERE usulogin IS NOT NULL
) AS src
ON tgt.usulogin = src.usulogin

WHEN MATCHED THEN
UPDATE
  SET tgt.usulogin = src.usulogin,
      tgt._update_at=CURRENT_TIMESTAMP()

WHEN NOT MATCHED THEN
  INSERT (usulogin)
  VALUES (src.usulogin);
```
