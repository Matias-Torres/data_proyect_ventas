```python
%sql
-- CREAR LA TABLA ventas_helados_bronze
DROP TABLE IF EXISTS catalog_ventas.raw.ventas_heladeria_bronze;

CREATE TABLE catalog_ventas.raw.ventas_heladeria_bronze
USING DELTA
AS
SELECT
  *,
  current_timestamp() AS _ingestion_ts,
  input_file_name() AS _source_file
FROM read_files(
  '/Volumes/catalog_ventas/raw/archivos/ventas_heladeria_bronze.csv',
  format => 'csv',
  header => true,
  sep => ';'
);
```
