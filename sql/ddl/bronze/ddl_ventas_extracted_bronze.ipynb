```python
%sql 
-- PASO 1 : CREAR CATALOGO
CREATE CATALOG IF NOT EXISTS catalog_ventas;

USE CATALOG catalog_ventas;
     
SHOW CATALOGS
     

```

```python
%sql
-- PASO 2: CREAR ESQUEMA RAW
CREATE SCHEMA IF NOT EXISTS catalog_ventas.raw
COMMENT 'Esquema para datos crudos sin procesar(Bronze Layer)';

SHOW SCHEMAS
```

```python
%sql
-- PASO 3: CREAR EL VOLUME PARA ARCHIVOS (ventas_helados_bronze.csv)
CREATE VOLUME IF NOT EXISTS catalog_ventas.raw.archivos
COMMENT 'Volume para almacenar archivos csv crudos';

SHOW VOLUMES IN catalog_ventas.raw
```

```python
%sql
-- PASO 4: VERIFICAR QUE SE SUBIO EL ARCHIVO CORRECTAMENTE
LIST '/Volumes/catalog_ventas/raw/archivos'
```

```python
%sql
-- PASO 5: CREAR LA TABLA ventas_helados_bronze
DROP TABLE IF EXISTS catalog_ventas.raw.ventas_helados_bronze;

CREATE TABLE IF NOT EXISTS catalog_ventas.raw.ventas_heladeria_bronze
  SELECT * FROM read_files(
    '/Volumes/catalog_ventas/raw/archivos/ventas_heladeria_bronze.csv',
    format => 'csv',
    header => true,
    sep => ';'

  )
```

```python
%sql
-- VERIFICAR TABLA CREADA
SHOW TABLES IN catalog_ventas.raw
```
