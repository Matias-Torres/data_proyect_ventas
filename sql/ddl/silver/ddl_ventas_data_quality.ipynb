```python
%sql
--CREAR ESQUEMA SILVER
USE CATALOG catalog_ventas;

CREATE SCHEMA IF NOT EXISTS catalog_ventas.silver
COMMENT 'Esquema para datos limpios y de calidad (Silver Layer)';

SHOW SCHEMAS
```

```python
%sql
--TABLA SILVER
DROP TABLE IF EXISTS catalog_ventas.silver.ventas_clean;

CREATE TABLE catalog_Ventas.silver.ventas_clean(

-- Identificador (auto-incremental)
id_venta BIGINT GENERATED ALWAYS AS IDENTITY (START WITH 1 INCREMENT BY 1) COMMENT ' Id unico de la venta (autogenerado)',

--Codigo de la venta del negocio
venta BIGINT COMMENT 'Codigo de venta',

--Caracteristicas del Articulo
articulo INT COMMENT 'Id unico del articulo',
descrip STRING COMMENT 'Descripcion del articulo',

--Caracteristicas principales
precio DECIMAL(15,2) COMMENT 'Precio del articulo',
cant INT COMMENT 'Cantidad de articulos vendidos',
total DECIMAL(15,2) COMMENT 'Total de la venta',

--Caracteristicas adicionales
usulogin STRING COMMENT 'Cajero que realizo la venta',
clientes STRING COMMENT 'Cliente',
vtaestado STRING COMMENT 'Estado de la venta (normal/anulada)',
vtafecha TIMESTAMP COMMENT 'Fecha de la venta',
condvtapos STRING COMMENT 'Metodos de pago',
delivery BOOLEAN COMMENT 'Envios a domicilio',
turno STRING COMMENT  'Turno de la venta (mañana/noche)',

--Metadata de procesamiento
_source_table STRING DEFAULT 'catalog_ventas.raw.ventas' COMMENT 'Tabla Origen',
_processing_timestamp TIMESTAMP DEFAULT CURRENT_TIMESTAMP() COMMENT 'Timestamp de procesamiento'

)
USING DELTA
TBLPROPERTIES ('delta.feature.allowColumnDefaults' = 'supported')
COMMENT' Ventas - Heladeria Datos limpios y validados (Silver Layer)'
```

```python
%sql
--VERIFICAR ESTRUCTURA 
DESCRIBE TABLE catalog_ventas.silver.ventas_clean
```
