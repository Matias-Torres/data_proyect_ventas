```python
%sql
DROP TABLE IF EXISTS catalog_ventas.gold.dim_fecha;

CREATE TABLE catalog_ventas.gold.dim_fecha(
    id_fecha BIGINT,
    vtafecha TIMESTAMP NOT NULL,
    hora INT,
    anio INT,
    mes INT,
    trimestre INT,
    dia_semana INT,
    es_fin_de_semana BOOLEAN,
    _created_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP()
)
USING DELTA
TBLPROPERTIES ('delta.feature.allowColumnDefaults' = 'supported')
COMMENT 'Tabla de dimension de fechas - Star Schema';
```
