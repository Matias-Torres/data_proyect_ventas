```python
%sql
DROP TABLE IF EXISTS catalog_ventas.gold.kpi_medios_pago;

CREATE OR REPLACE TABLE catalog_ventas.gold.kpi_medios_pago
USING DELTA
AS

SELECT
    mp.medio_pago,
    COUNT(DISTINCT f.id_venta) AS tickets,
    SUM(f.total)               AS facturacion,
    ROUND(SUM(f.total) * 100.0 /
          SUM(SUM(f.total)) OVER (),2) AS pct_facturacion

FROM catalog_ventas.gold.fact_ventas f

LEFT JOIN catalog_ventas.gold.dim_medio_pago mp
  ON f.id_medio_pago = mp.id_medio_pago

GROUP BY mp.medio_pago
ORDER BY facturacion DESC
```
