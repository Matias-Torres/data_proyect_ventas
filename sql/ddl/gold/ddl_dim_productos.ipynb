```python
%sql
DROP TABLE IF EXISTS catalog_ventas.gold.dim_producto;

CREATE TABLE catalog_ventas.gold.dim_producto(
  id_articulo BIGINT GENERATED ALWAYS AS IDENTITY (START WITH 1 INCREMENT BY 1),
  articulo INT,
  descrip STRING,
  _created_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP(),
  _updated_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP()
)
USING DELTA
TBLPROPERTIES ('delta.feature.allowColumnDefaults' = 'supported')
COMMENT 'Tabla de dimension de articulos - Star Schema';

```
