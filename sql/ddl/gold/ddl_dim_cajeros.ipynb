```python
%sql
DROP TABLE IF EXISTS catalog_ventas.gold.dim_cajero;

CREATE TABLE catalog_ventas.gold.dim_cajero(
  id_usulogin BIGINT GENERATED ALWAYS AS IDENTITY (START WITH 1 INCREMENT BY 1),
  usulogin STRING NOT NULL,
  turno STRING,
  _created_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP(),
  _updated_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP()
)
USING DELTA
TBLPROPERTIES ('delta.feature.allowColumnDefaults' = 'supported')
COMMENT 'Tabla de dimension de cajeros - Star Schema';
```
