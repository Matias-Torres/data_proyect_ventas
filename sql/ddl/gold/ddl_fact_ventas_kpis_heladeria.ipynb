```python
%sql

DROP TABLE IF EXISTS catalog_ventas.gold.kpis_heladeria;

CREATE TABLE catalog_ventas.gold.kpis_heladeria
USING DELTA
COMMENT 'KPIS del Mercado Heladeria'
AS
WITH kpis AS (
SELECT
  COUNT(*) AS total_ventas,
  COUNT(DISTINCT articulo) AS total_productos,
  
  SUM(total) AS facturado,
  SUM(cant) AS unidades,

  COUNT(CASE WHEN clientes='socio' THEN 1 END) AS total_clientes_socio,
  COUNT(CASE WHEN clientes='no_socio' THEN 1 END) AS total_clientes_no_socio,
  
  COUNT(CASE WHEN condvtapos='credito' THEN 1 END) AS total_credito,
  COUNT(CASE WHEN condvtapos='efectivo' THEN 1 END) AS total_efectivo,
  COUNT(CASE WHEN condvtapos='debito' THEN 1 END) AS total_debito,
  
  COUNT(CASE WHEN delivery=true THEN 1 END) AS total_delivery,
  COUNT(CASE WHEN delivery=false THEN 1 END) AS total_local
  

  FROM catalog_ventas.silver.ventas_clean
  
)
SELECT
  total_ventas,
  total_productos,
  facturado,
  unidades,
  ROUND(total_clientes_socio * 100.00 / total_ventas,2 ) AS pct_clientes_socio,
  ROUND(total_clientes_no_socio * 100.00 / total_ventas,2 ) AS pct_clientes_no_socio,
  ROUND(total_local * 100.00 / total_ventas,2 ) AS pct_local,
  ROUND(total_delivery * 100.00 / total_ventas,2 ) AS pct_delivery,
  ROUND(total_credito * 100.00 / total_ventas,2 ) AS pct_credito,
  ROUND(total_efectivo * 100.00 / total_ventas,2 ) AS pct_efectivo,
  ROUND(total_debito * 100.00 / total_ventas,2 ) AS pct_debito,
  CURRENT_TIMESTAMP() as _refresh_timestamp
FROM kpis
```
