```python
%sql
DROP TABLE IF EXISTS catalog_ventas.gold.dim_medio_pago;

CREATE TABLE catalog_ventas.gold.dim_medio_pago(
  id_medio_pago BIGINT GENERATED ALWAYS AS IDENTITY (START WITH 1 INCREMENT BY 1),
  medio_pago STRING,
  _created_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP()
)
USING DELTA
TBLPROPERTIES ('delta.feature.allowColumnDefaults' = 'supported')
COMMENT 'Tabla de dimension de medios de pagos (Debito,Credito,Efectivo) - Star Schema';
```
