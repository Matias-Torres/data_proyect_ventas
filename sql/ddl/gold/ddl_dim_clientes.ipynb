```python
%sql
DROP TABLE IF EXISTS catalog_ventas.gold.dim_clientes;

CREATE TABLE catalog_ventas.gold.dim_clientes(
  id_cliente BIGINT GENERATED ALWAYS AS IDENTITY (START WITH 1 INCREMENT BY 1),
  cliente STRING,
  _created_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP(),
  _updated_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP()
)
USING DELTA
TBLPROPERTIES ('delta.feature.allowColumnDefaults' = 'supported')
COMMENT 'Tabla de dimension de clientes - Star Schema';
```
