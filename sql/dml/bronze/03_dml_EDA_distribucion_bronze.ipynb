## **EDA - Cardinalidad y Distribución de Categóricos**


```python
%sql
-- DISTRIBUCION DE TURNO
SELECT
    turno,
    COUNT(*) AS total,
    ROUND(COUNT(*) * 100.0 / SUM(COUNT(*)) OVER(),2) AS pct_turno
FROM catalog_ventas.raw.ventas_heladeria_bronze
    GROUP BY turno
    ORDER BY total DESC

/*
Muy alta cardinalidad, muchos valores distintos.
Muy poco turnos concentran muchisimas ventas
muchos turnos con pocas ventas bajas  
*/
```

```python
%sql
-- DISTRIBUCION DE VTAOPERACION
SELECT
    vtaoperacion,
    COUNT(*) AS total,
    ROUND(COUNT(*) * 100.0 / SUM(COUNT(*)) OVER(),2) AS pct_vtaoperacion
FROM catalog_ventas.raw.ventas_heladeria_bronze
    GROUP BY vtaoperacion
    ORDER BY total DESC

/*
VL: venta a socios 
VF: venta a no socios
VG: venta gastronomica
CL: otros
*/
```

```python
%sql

-- DISTRIBUCION DE VTAESTADO
SELECT
    vtaestado,
    COUNT(*) AS total,
    ROUND(COUNT(*) * 100.0 / SUM(COUNT(*)) OVER(),2) AS pct_vtaestado
FROM catalog_ventas.raw.ventas_heladeria_bronze
    GROUP BY vtaestado
    ORDER BY total DESC


/*
Baja cardinalidad, solo 2 valores distintos
Distribucion altamente cocnentrada en ventas normales
Distribucion baja en ventas anuladas (saludable)
*/
```

```python
%sql

-- DISTRIBUCION DE CLINOMBRE
SELECT
    clinombre,
    COUNT(*) AS total,
    ROUND(COUNT(*) * 100.0 / SUM(COUNT(*)) OVER(),2) AS pct_clinombre
FROM catalog_ventas.raw.ventas_heladeria_bronze
    GROUP BY clinombre
    ORDER BY total DESC

/*
El 95% corresponde a la venta minorista de helados en sucursales.
El 5% restante corresponde a ventas a comerciantes o otros clientes.
*/
```

```python
%sql
-- DISTRIBUCION DE CONDVTAPOS
SELECT
    condvtapos,
    COUNT(*) AS total,
    ROUND(COUNT(*) * 100.0 / SUM(COUNT(*)) OVER(),2) AS pct_condvtapos
FROM catalog_ventas.raw.ventas_heladeria_bronze
    GROUP BY condvtapos
    ORDER BY total DESC

/*
TD: tarjeta debito
EF: efectivo
TC: tarjeta credito
MU: multiples opciones de pagos
XX: cancelados
*/
```

```python
%sql
-- DISTRIBUCION DE DELIVERY
SELECT
    delivery,
    COUNT(*) AS total,
    ROUND(COUNT(*) * 100.0 / SUM(COUNT(*)) OVER(),2) AS pct_delivery
FROM catalog_ventas.raw.ventas_heladeria_bronze
    GROUP BY delivery
    ORDER BY total DESC
```
