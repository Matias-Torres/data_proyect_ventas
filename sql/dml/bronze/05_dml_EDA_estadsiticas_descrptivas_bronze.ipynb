## **EDA - Estadísticas Descriptivas**


```python
%sql
-- ESTADISTICA DE PRECIO SEGUN PRODUCTOS
SELECT
  descrip,
  COUNT(*) AS total_ventas,
  ROUND(MIN(precio),2) AS precio_minimo,
  ROUND(MAX(precio),2) AS precio_maximo,
  ROUND(AVG(precio),2) AS precio_promedio,
  ROUND(PERCENTILE(precio,0.25),2)  AS precio_mediana,
  ROUND(PERCENTILE(precio,0.5),2)  AS precio_50,
  ROUND(PERCENTILE(precio,0.75),2) AS precio_75,
  ROUND(PERCENTILE(precio,0.9),2) AS precio_90
FROM ventas_clean
WHERE vtaestado='normal'
 AND precio IS NOT NULL 
 AND precio > 0
 AND precio != 0.01
GROUP BY descrip
ORDER BY total_ventas DESC
```

```python
%sql
-- ESTADISTCA DE CANTIDAD
SELECT
  COUNT(*) AS total_ventas,
  ROUND(MIN(cant),2) AS cant_minimo,
  ROUND(MAX(cant),2) AS cant_maximo,
  ROUND(AVG(cant),2) AS cant_promedio,
  ROUND(PERCENTILE(cant,0.25),2)  AS cant_mediana,
  ROUND(PERCENTILE(cant,0.5),2)  AS cant_50,
  ROUND(PERCENTILE(cant,0.75),2) AS cant_75,
  ROUND(PERCENTILE(cant,0.9),2) AS cant_90
FROM ventas_clean
WHERE vtaestado='normal'
```

```python
%sql
-- ESTADISTICA DE INGRESOS 
SELECT
  COUNT(*) AS total_ventas,
  ROUND(SUM(total),2) AS ingreso_total,
  ROUND(MIN(total),2) AS total_minimo,
  ROUND(MAX(total),2) AS total_maximo,
  ROUND(AVG(total),2) AS total_promedio,
  ROUND(PERCENTILE(total,0.25),2)  AS total_mediana,
  ROUND(PERCENTILE(total,0.5),2)  AS total_50,
  ROUND(PERCENTILE(total,0.75),2) AS total_75,
  ROUND(PERCENTILE(total,0.9),2) AS total_90
FROM ventas_clean
WHERE vtaestado='normal'
 AND total IS NOT NULL 
 AND precio > 0
 AND total != 0.01
ORDER BY ingreso_total DESC
```
