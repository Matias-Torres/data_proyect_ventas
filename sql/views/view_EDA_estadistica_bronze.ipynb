```python
%sql
--VISTA TEMPORAL PARA CALCULAR ESTADISTICAS SIN PROBLEMAS DE FORMATOS
DROP VIEW IF EXISTS ventas_clean;
CREATE OR REPLACE  VIEW ventas_clean AS
SELECT
    venta,
    articulo,
    LOWER(TRIM(descrip)) AS descrip,
    TRY_CAST(
        REGEXP_REPLACE(precio, ',', '.') AS DOUBLE
    ) AS precio,

    TRY_CAST(cant AS INT) AS cant,

    TRY_CAST(
        REGEXP_REPLACE(total, ',', '.') AS DOUBLE
    ) AS total,
    LOWER(TRIM(usulogin)) AS usulogin,
    LOWER(TRIM(clinombre)) AS clinombre,
    cliente,
    vtaoperacion,
    LOWER(TRIM(vtaestado)) AS vtaestado,
    TO_TIMESTAMP(vtafecha, 'd/M/yyyy HH:mm') AS vtafecha,
    condvtapos,
    UPPER(TRIM(delivery)) AS delivery,
    turno,
    caja,
    comprobante,
    LOWER(TRIM(succodigo)) AS succodigo,
    sucursal   
FROM catalog_ventas.raw.ventas_heladeria_bronze;
```
